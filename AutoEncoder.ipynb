- It is the neural network that learn to copy from input to output.
- If u see the X is the input and x' is the output.
- Autoencoder is divided into two parts first is the encoder and second is the decoder.
- Encoder is mapping down input space to the lower dimensions latent space.
- Here Z latent space and it is called as bottle neck layer.
- This Z is also called as lower dimensional space.
- It lower dimensional data which is completely in unsupervised mannner.
- Autoencoder is unsupervised learning technique..we do not required label data.
- It is more powerful as compare to the PCA.
- It also help to get the information at the output with less significantly loss in information than PCA in non-linear space.
- PCA is going to work linear dataset and Autoencoder work on non-linear dataset.
- Encoder takes input data and convert into the lower dimensional space.
- Decoder is helping to get the lower dimensional data and convert that data into reconstructed vector space.
- Where the dimensionality of X' is getting approximately to the dimensionality of X.
- In the case of image,if any image is getting compress in encoder and use the decoder which is helping to reconstruct the image but not completely but as approximately.
- In the image compression there will be lossy compression over there.
- The error inbetween the input and output vector called reconstruction error.
- One potential use case in autoencoder is called as anomaly detection.
- Training objective of the autoencoder is to minimizes the reconstruction error.
- As the training progresses the weight of each neuron is getting updated.
- Here in figure the encode is downsampler and decoder is the upsampler.
- Encoder and decoder can be our convolutional neural network can be feedforeward neural network or can be our RNN neural network.
